```python
import pandas as pd
from collections import Counter
import ast
from google.colab import drive

drive.mount('/content/drive')


df = pd.read_csv('/content/drive/MyDrive/ThesisData/final_enriched_news_unique.csv')

def count_top_entities(df, label='ORG', top_n=10):
    all_ents = []
    for row in df['entities'].dropna():
        parts = [e.strip() for e in row.split(',')]
        for p in parts:
            if f"({label})" in p:
                name = p.replace(f"({label})", "").strip()
                all_ents.append(name)

    return Counter(all_ents).most_common(top_n)


top_orgs = count_top_entities(df, 'ORG')
top_people = count_top_entities(df, 'PERSON')

print("--- TOP 10 ORGANIZATIONS ---")
for org, count in top_orgs:
    print(f"{org}: {count}")

print("\n--- TOP 10 PEOPLE ---")
for person, count in top_people:
    print(f"{person}: {count}")
```
