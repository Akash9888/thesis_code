```python
import pandas as pd
from google.colab import drive

drive.mount('/content/drive')
df = pd.read_csv('/content/drive/MyDrive/ThesisData/final_enriched_news_unique.csv')

ticker_map = {
    'Pfizer': 'PFE',
    'Eli Lilly': 'LLY',
    'Johnson & Johnson': 'JNJ',
    'J&J': 'JNJ',
    'Merck': 'MRK',
    'AbbVie': 'ABBV',
    'AstraZeneca': 'AZN',
    'Novo Nordisk': 'NVO',
    'Moderna': 'MRNA',
    'Gilead': 'GILD',
    'Amgen': 'AMGN',
    'FDA': 'XLV',        # Maps regulatory news to the Sector ETF
    'ObamaCare': 'XLV',  # Maps policy news to the Sector ETF
    'Medicare': 'XLV'
}

def map_to_ticker(entities):
    if pd.isna(entities): return 'UNKNOWN'
    for entity_name, ticker in ticker_map.items():
        if entity_name.lower() in str(entities).lower():
            return ticker
    return 'CONTEXT_ONLY'


df['ticker'] = df['entities'].apply(map_to_ticker)

df.to_csv('/content/drive/MyDrive/ThesisData/news_with_tickers.csv', index=False)
print("Ticker mapping complete")
```
