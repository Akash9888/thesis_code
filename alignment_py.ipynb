```python
import pandas as pd
from google.colab import drive

drive.mount('/content/drive')


news_df = pd.read_csv('/content/drive/MyDrive/ThesisData/news_with_tickers.csv')
stocks_df = pd.read_csv('/content/drive/MyDrive/ThesisData/historical_prices.csv')

news_df = news_df.rename(columns={'date': 'Date'})

news_df = news_df.rename(columns={'ticker': 'Ticker'})

news_df['Date'] = pd.to_datetime(news_df['Date'], format='mixed')
stocks_df['Date'] = pd.to_datetime(stocks_df['Date'], format='mixed')


print("Aligning news sentiment with stock prices (including weekends)...")

news_df = news_df.sort_values('Date')
stocks_df = stocks_df.sort_values('Date')

final_df_list = []
for ticker in stocks_df['Ticker'].unique():
    ticker_news = news_df[news_df['Ticker'] == ticker]
    ticker_stocks = stocks_df[stocks_df['Ticker'] == ticker]

    temp_merged = pd.merge_asof(
        ticker_news,
        ticker_stocks,
        on='Date',
        by='Ticker',
        direction='forward'
    )
    final_df_list.append(temp_merged)


final_master_df = pd.concat(final_df_list)
final_master_df = final_master_df.dropna(subset=['Daily_Return'])


OUTPUT_PATH = '/content/drive/MyDrive/ThesisData/MASTER_THESIS_DATA.csv'
final_master_df.to_csv(OUTPUT_PATH, index=False)

print("\n" + "="*30)
print(f"SUCCESS: Master Dataset created with {len(final_master_df)} rows.")
print(f"File Saved: {OUTPUT_PATH}")
print(final_master_df.tail())
print("="*30)
```
