```python

import yfinance as yf
import pandas as pd
import os
from google.colab import drive

drive.mount('/content/drive')


OUTPUT_FILE = '/content/drive/MyDrive/ThesisData/historical_prices.csv'
START_DATE = "2016-01-01"
END_DATE = "2021-01-05"


TICKERS = ["PFE", "LLY", "MRNA", "JNJ", "NVO", "MRK", "GILD", "ABBV", "XLV"]

def fetch_and_flatten_stocks():
    print(f"Downloading {len(TICKERS)} tickers...")


    data = yf.download(TICKERS, start=START_DATE, end=END_DATE, auto_adjust=True)


    df_long = data['Close'].stack().reset_index()
    df_long.columns = ['Date', 'Ticker', 'Close']
    df_long['Date'] = pd.to_datetime(df_long['Date']).dt.date


    df_long = df_long.sort_values(['Ticker', 'Date'])
    df_long['Daily_Return'] = df_long.groupby('Ticker')['Close'].pct_change()


    df_long.to_csv(OUTPUT_FILE, index=False)

    print("\n" + "="*30)
    print(f"Total Rows Saved: {len(df_long)}")
    print("Preview of Data:")
    print(df_long.tail())
    print("="*30)

if __name__ == "__main__":
    fetch_and_flatten_stocks()
```
